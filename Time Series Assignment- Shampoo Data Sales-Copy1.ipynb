# ARIMA Model

To determine the values of p and q in ARIMA model, we will plot the autocorrelation plots.

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
data = data.rename(columns= {'Sales of shampoo over a three year period': 'sales'})
```

```python
#lag plots 
lag_acf = acf(data.sales.dropna(), nlags=20, fft=True)
ACF = pd.Series(lag_acf)

lag_pacf = pacf(data.sales.dropna(), nlags=15, method = 'ols')
PACF = pd.Series(lag_pacf)

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (16, 16))
g1 = ACF.plot(ax = axes[0,0])
ix = np.argmin(ACF >= 0)
g1.scatter(x = ix, y = ACF[ix], color = 'black')
axes[0][0].set_title('AutoCorrelation function plot')

f1 = ACF.plot(ax = axes[0,1], kind = 'bar')
f1.axhline(y = ACF[5], ls = '-.', color = 'black')
f1.axvline(x = 5, ls = '-.', color = 'black')
axes[0][1].set_title('AutoCorrelation function barplot')

g2 = PACF.plot(ax = axes[1,0])
ix = np.argmin(PACF >= 0)
g2.scatter(x = ix, y = PACF[ix], color = 'black')
axes[1][0].set_title('Partial AutoCorrelation function plot')

f2 = PACF.plot(ax = axes[1,1], kind = 'bar')
f2.axhline(y = ACF[5], ls = '-.', color = 'black')
f2.axvline(x = 5, ls = '-.', color = 'black')
axes[1][1].set_title('Partial AutoCorrelation function barplot')

```

# Optimizing ARIMA Model

```python
def evaluate_model(data, model_order):
    size = int(len(data['sales'])*0.66)
    X = data.values
    train, test = X[0:size], X[size:len(X)]
    history = [x for x in train]
    predictions = list()
    
    
    for t in range(len(test)):
        model = ARIMA(history, order = model_order)
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        pred = output[0]
        predictions.append(pred)
        actual = test[t]
        history.append(actual)
    
    return(mean_squared_error(test, predictions))
```

```python
p_values = range(2, 12,2)
d_values = range(0, 3)
q_values = range(0, 3)
```

```python
def evaluate_order(data, p_values,d_values,q_values):
    best_mse, best_order = float('inf'), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = evaluate_model(data=data, model_order = order)
                    if mse < best_mse:
                        best_order = order
                        best_mse = mse
                    print('ARIMA%s MSE=%.3f' % (order,mse))
                except:
                    continue
    print('Best ARIMA%s Best MSE%.3f' %(best_order, best_mse))

    
                 
```
